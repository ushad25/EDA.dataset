```python



# Count players by country
most_players_country = df['Nationality'].value_counts().idxmax()
print("Country with the most players:", most_players_country)
```

```python
Data Story -
The analysis revealed that the country with the most players in FIFA is "England".
This suggests that England has a significant representation of players in the FIFA dataset.
```

```python
2. Plot a bar chart of 5 top countries with the most number of players.
To visualize the top 5 countries with the most players, we create a bar chart showing the player counts for each of these countries.
```

```python


# Top 5 countries with the most players
top_countries = df['Nationality'].value_counts().head(5)

# Plotting the bar chart
top_countries.plot(kind='bar', title='Top 5 Countries with Most Players')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.show()
```

```python
Data Story -
The bar chart illustrates the top 5 countries with the highest number of players in FIFA. 
It shows that England, Germany, Spain, Argentina, and France are the leading countries in terms of player representation.
```

```python
 Which player has the highest salary?
```

```python
# Identify player with the highest salary
highest_salary_player = df.loc[df['Wage'].idxmax()]['Name']
print("Player with the highest salary:", highest_salary_player)
```

```python
Data Story -
The analysis reveals that the player with the highest salary in FIFA is  F. Marchetti
```

```python
Plot a histogram to get the salary range of the players.
```

```python
# Plotting the histogram
plt.hist(df['Wage'], bins=20, color='skyblue', edgecolor='black')
plt.title('Salary Range of Players')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
Data Story -
The histogram provides insights into the distribution of player salaries. 
It shows the frequency of different salary ranges among FIFA players, indicating the most common salary levels.


```

```python
# Identify the club with the most players
most_players_club = df['Club'].value_counts().idxmax()
print("Club with the most players:", most_players_club)
```

```python
 Which foot is most preferred by the players? Draw a bar chart for preferred foot.
```

```python
# Count preferred foot
preferred_foot = df['Preferred Foot'].value_counts()

# Plotting the bar chart
preferred_foot.plot(kind='bar', title='Preferred Foot of Players')
plt.xlabel('Preferred Foot')
plt.ylabel('Number of Players')
plt.show()
```

```python
Data Story -
The bar chart illustrates the distribution of preferred foot among FIFA players, indicating the most common foot preference among them.
```

```python
# - Age distribution of players
plt.hist(df['Age'], bins=20, color='orange', edgecolor='black')
plt.title('Age Distribution of Players')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['Overall'])
plt.title('Box Plot of Overall Ratings')
plt.ylabel('Overall Rating')
plt.show()
```
